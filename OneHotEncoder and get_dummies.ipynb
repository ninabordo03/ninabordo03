```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('penguins_size.csv')
df
```

```python
df.isna().sum()
```

```python
df.info()
```

```python
df = df.dropna()
```

```python
print(df['species'].unique())
print(df['island'].unique())
print(df['sex'].unique())
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder()
ohe
```

```python
feat = ohe.fit_transform(df[['species', 'island', 'sex']]).toarray()
feat
```

```python
cat = ohe.categories_
cat
```

```python
cat = np.array(cat).ravel()
cat
```

```python
features = pd.DataFrame(feat, columns = cat)
features
```

```python
dfn = df.drop(['species', 'island', 'sex'], axis = 1)
dfn
```

```python
print(dfn.shape)
print(features.shape)
```

```python
con = pd.concat([dfn, features], axis = 1)
con
```
